```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
rows = ['1', '2', '3', '4', '5']
```

```python
def plotMNB():
    plt.figure(dpi=1200)
    plt.plot(rows, [0.7911, 0.7939, 0.8189, 0.7855,
             0.8106], color="C0", linewidth="1", label="count", marker="s")
    plt.plot(rows, [0.7911, 0.8022, 0.8134, 0.7911,
             0.8134], color="C1", linewidth="1.3", label="TF-IDF", linestyle='dashdot', marker="D")
    plt.plot(rows, [0.7855, 0.7855, 0.8134, 0.7911,
             0.8050], color="C2", linewidth="1.3", label="binary", linestyle='dashed', marker="o")

    plt.title('Multinomial Naive Bayes')
    plt.legend()
    plt.ylabel('accuracy')
    plt.xlabel('preprocessing row')
    plt.show()


def plotSVM():
    plt.figure(dpi=1200)
    plt.plot(rows, [0.8078, 0.8078, 0.8106, 0.7994,
             0.7967], color="C0", linewidth="1", label="count", marker="s")
    plt.plot(rows, [0.7911, 0.7911, 0.7994, 0.7883,
             0.8050], color="C1", linewidth="1.3", label="TF-IDF", linestyle='dashdot', marker="D")
    plt.plot(rows, [0.8022, 0.8078, 0.7994, 0.8078,
             0.7939], color="C2", linewidth="1.3", label="binary", linestyle='dashed', marker="o")

    plt.title('SVM')
    plt.legend()
    plt.ylabel('accuracy')
    plt.xlabel('preprocessing row')
    plt.show()


def plotLR():
    plt.figure(dpi=1200)
    plt.plot(rows, [0.8078, 0.8022, 0.8078, 0.7994,
             0.7967], color="C0", linewidth="1", label="count", marker="s")
    plt.plot(rows, [0.7911, 0.7994, 0.7967, 0.7883,
             0.7967], color="C1", linewidth="1.3", label="TF-IDF", linestyle='dashdot', marker="D")
    plt.plot(rows, [0.7994, 0.8022, 0.8050, 0.7994, 0.7883],
             color="C2", linewidth="1.3", label="binary", linestyle='dashed', marker="o")

    plt.title('Logistic Regression')
    plt.legend()
    plt.ylabel('accuracy')
    plt.xlabel('preprocessing row')
    plt.show()
```

```python
plotMNB()
```

```python
plotSVM()
```

```python
plotLR()
```
