```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
hist_eps_01 = pd.read_csv("models/histories/history_dp-eps=0-1.csv")
hist_eps_1 = pd.read_csv("models/histories/history_dp-eps=1.csv")
hist_eps_10 = pd.read_csv("models/histories/history_dp-eps=10.csv")
hist_eps_baseline = pd.read_csv("models/histories/history_count_row_3.csv")
```

```python
plt.figure(dpi=1200)

plt.plot(range(1, len(hist_eps_01.val_accuracy)+1),
         hist_eps_01.val_accuracy, color="C1", linewidth="1", label=" = 0.1")
plt.plot(range(1, len(hist_eps_1.val_accuracy)+1),
         hist_eps_1.val_accuracy, color="C0", linewidth="1", label=" = 1")
plt.plot(range(1, len(hist_eps_10.val_accuracy)+1),
         hist_eps_10.val_accuracy, color="C2", linewidth="1", label=" = 10")
plt.plot(range(1, len(hist_eps_baseline.val_accuracy)+1),
         hist_eps_baseline.val_accuracy, color="C3", linewidth="1", label=" = ")

plt.title('DP-Models')
plt.legend()
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```
